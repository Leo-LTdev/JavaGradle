* Struture du projet 
  

  Class personnage
  
  1. variable 
     - vie                  int
     - speed              int
     - armor               int
     - attack                  int
     - exp                  double
  
  1. Méthode personnage 
     - [ ] vie (get, set)
     - [ ] speed (get, set)
     - [ ] armor (get, set)
     - [ ] attack (get, set)
     - [ ] exp (get, set)


  Class Aventurier sous class personnage

  1. variable 
     - inventaire           array ( item )
     - level                int 
     - équipement           array ( item )
  
  2. Méthode Aventurier
     - [ ] inventaire (set, get, add, remove)
     - [ ] level (get, set, levelUp)
     - equip()
     - unequip()
     - has weapon()

  Class Monstre

     - vie

  2. Méthode de Monstre
     - [ ] vie (get, set)
     - [ ] type (get)
     - [ ] speed (get, set)
     - [ ] armor (get, set)
     - [ ] attack (get, set)

  Class object
  1. variable 
     - durabilité
  
  2. Méthode personnage 
     - [ ] durabilité (get, set, durabilityLost, repair/refill)
  

  Class Arme sous class de object
  1. variable 
     - atk 
     - type (contondant, tranchant, perforant)
     - level
  
  2. Méthode Arme 
     - [ ] atk (set, get)
     - [ ] type (get)
     - [ ] level (get)

  
  Class Armure sous class de object

     - armor
     - type (tissu, cuir, acier)

  2. Méthode Armure 
     - [ ] armor (set, get)
     - [ ] type (get)
     - [ ] level (get)

```java
import java.util.ArrayList;

public class Character {
    private int Life;
    private int Speed;
    private float Armor;
    private float Attack;
    private float EXP;

    public Character(int Life,int Speed,float Armor,float Attack,float EXP) {
        this.Life = Life;
        this.Speed = Speed;
        this.Armor = Armor;
        this.Attack = Attack;
        this.EXP = EXP;
    }

    public int getLife() {
        return Life;
    }
    public int getSpeed() {
        return Speed;
    }
    public float getArmor() {
        return Armor;
    }
    public float getAttack() {
        return Attack;
    }
    public float getEXP() {
        return EXP;
    }

    public int Attack() {
        return 10;
    }
}

public class Adventurer extends Character {
    private ArrayList<String> Inventory;

    public Adventurer(int Life, int Speed, float Armor, float Attack, float EXP) {
        super(Life,Speed,Armor,Attack,EXP);
        this.Inventory = new ArrayList<String>();
    }

    public ArrayList<String> getInv() {
        return Inventory;
    }
}

public class Human extends Adventurer {
    public Human() {
        super(110, 10, 0, 10, 0);
    }
}

public class Elves extends Adventurer {
    public Elves() {
        super(80, 20, 0, 15, 0);
    }
}

public class Monsters {

    protected String Type;
    protected int Level;
    protected int Life;
    protected int Speed;
    protected double Armor;
    protected double Attack;
    protected double Xp;

    public Monsters(String Type, int Level, int Life, int Speed, double Armor, double Attack, double Xp) {
        this.Type = Type;
        this.Level = Level;
        this.Life = Life;
        this.Speed = Speed;
        this.Armor = Armor;
        this.Attack = Attack;
        this.Xp = Xp;
        
    }
    
    public void ReceiveDMG(int dmg) {
        this.Life -= dmg;
    }

}
```

```java
public class Tyrannides extends Monsters {

    private int Claws;


    public Tyrannides(String Type, int Level, int Life, int Speed, double Armor, double Attack, double Xp, int Claws) {
        super("Tyrannides", 1, 50, 10, 0.5, 5.5, 2.5);
        this.Claws = Claws;
        
    }
    
    public int getClaws() {
        return Claws;
    }
} 
```

```java
public class Undead extends Monsters {

    private int ReanimationRate;

    public Undead(String Type, int Level, int Life, int Speed, double Armor, double Attack, double Xp, int ReanimationRate) {
        super("Undead", 1, 40, 5, 0.5, 5.5, 2.5);
        this.ReanimationRate = ReanimationRate;
    }

    public int getReanimationRate() {
        return ReanimationRate;
    }

}
```

```java
public class Orcs extends Monsters {

    private String Tribe;

    public Orcs() {
        super("Orcs", 20, 200, 3, 0.5, 5.5, 2.5);
        this.Tribe = Tribe;
    }

    public String getTribe() {
        return Tribe;
    }
}
```

```java
Scanner scanner = new Scanner(System.in);
Adventurer Player = null;
boolean inGame = true;

System.out.println("Choisis une race parmis celle-ci :");
System.out.println("1 : Humain");
System.out.println("2 : Elfes");

if (scanner.hasNextInt()) {
    int nombre = scanner.nextInt();
    if (nombre == 1) {
        Player = new Human(); 
    } 
    if (nombre == 2) {
        Player = new Elves(); 
    } 
}

while (inGame && Player != null) {
    System.out.println("1 : Voir ses stats");
    System.out.println("2 : Baston");
    System.out.println("9 : Quitter la partie");
    if (scanner.hasNextInt()) {
        int nombre = scanner.nextInt();
        if (nombre == 1) {
            System.out.println("Vie : " + Player.getLife());
            System.out.println("Armure : " + Player.getArmor());
            System.out.println("Vitesse : " + Player.getSpeed());
            System.out.println("Attaque : " + Player.getAttack());
            System.out.println("EXP : " + Player.getEXP());
        } 
        if (nombre == 2) {
            while (Player.getLife() > 100) {
                Orcs Monster = new Orcs();
                System.out.println("1 : Attaquer");
                int nombre = scanner.nextInt();
                if (nombre == 1) {  
                    Monster.ReceiveDMG(Player.Attack());
                }
            }
        } 
        if (nombre == 9) {
            inGame = false;
            System.out.println("Fin du jeu.");
        }
    }
}
```
